```python
import pandas as pd

file_path = "../farmers-protest-tweets-2021-2-4.json"
df = pd.read_json(file_path, lines=True)

df['date_t'] = pd.to_datetime(df['date']).dt.date

max_days = df.groupby("date_t").size().nlargest(10)

df_usernames = pd.concat([df[['date_t']], df['user'].str['username']], axis = 1)

max_users = df_usernames.groupby(["date_t","user"]).size()

most_repeated_ids_per_day = []
for date in max_days.index:
    most_repeated_ids_per_day.append((date, max_users[date].idxmax())) 
most_repeated_ids_per_day
```

```python
from emoji import UNICODE_EMOJI
import json

file_path = "../farmers-protest-tweets-2021-2-4.json"

emojis = {}
with open(file_path) as f:
        for jsonObj in f:
            tweetDict = json.loads(jsonObj)
            emojis = [c for c in tweetDict['content'] if c in UNICODE_EMOJI]
            print(emojis)
            for word in tweetDict['content']:
                emojis = [c for c in word if c in UNICODE_EMOJI]
                print(emojis)
                if word in UNICODE_EMOJI:
                    if emojis.get(word) is not None:
                        emojis[word] = emojis[word] + 1
                        break
                    else:
                        emojis[word] = 1

# Sort the dictionary by values in descending order
emojis = sorted(emojis.items(), key=lambda x: x[1], reverse=True)

# Extract the top 10 values
emojis = emojis[:10]
emojis
```

```python
import re
import json

file_path = "../farmers-protest-tweets-2021-2-4.json"
emoji_pattern = re.compile(r'[\U0001F300-\U0001F6FF\U0001F900-\U0001F9FF\u2600-\u26FF\u2700-\u27BF]', flags=re.UNICODE)

emojis_dict = {}
with open(file_path) as f:
        for jsonObj in f:
            tweetDict = json.loads(jsonObj)
            emojis = emoji_pattern.findall(tweetDict['content'])
            for emoji in emojis:
                if emojis_dict.get(emoji) is not None:
                    emojis_dict[emoji] = emojis_dict[emoji] + 1
                else:
                    emojis_dict[emoji] = 1


# Sort the dictionary by values in descending order
emojis_dict = sorted(emojis_dict.items(), key=lambda x: x[1], reverse=True)

# Extract the top 10 values
emojis_dict = emojis_dict[:10]
emojis_dict
```

```python
import json

file_path = "../farmers-protest-tweets-2021-2-4.json"

mentioned_users_dict = {}
with open(file_path) as f:
        for jsonObj in f:
            tweetDict = json.loads(jsonObj)
            mentioned_users = tweetDict['mentionedUsers']
            if mentioned_users is not None:
                for user in mentioned_users:
                    username = user['username']
                    if mentioned_users_dict.get(username) is not None:
                        mentioned_users_dict[username] = mentioned_users_dict[username] + 1
                    else:
                        mentioned_users_dict[username] = 1


# Sort the dictionary by values in descending order
mentioned_users_dict = sorted(mentioned_users_dict.items(), key=lambda x: x[1], reverse=True)

# Extract the top 10 values
mentioned_users_dict = mentioned_users_dict[:10]
mentioned_users_dict
```
